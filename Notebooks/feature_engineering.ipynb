Latitude and longitude were imputed using ZIP-code level centroids computed from 287,134 records with valid geolocation data. Remaining missing values were filled using global medians.

```python
zip_coords = (
    sales_pluto_combined
    .groupby("ZIP CODE")[["latitude","longitude"]]
    .mean()
    .reset_index()
)

lat_map = zip_coords.set_index("ZIP CODE")["latitude"].to_dict()
lon_map = zip_coords.set_index("ZIP CODE")["longitude"].to_dict()
```

```python
mask = sales_pluto_combined["latitude"].isna()

sales_pluto_combined.loc[mask, "latitude"] = (
    sales_pluto_combined.loc[mask, "ZIP CODE"].map(lat_map)
)

sales_pluto_combined.loc[mask, "longitude"] = (
    sales_pluto_combined.loc[mask, "ZIP CODE"].map(lon_map)
)
```

```python
sales_pluto_combined.info()
```

```python
sales_pluto_combined[sales_pluto_combined['latitude'].isnull() | sales_pluto_combined['longitude'].isnull()]
```

For 63 entries, latitude and longitude is still missing. Borough level centroid was tried for imputation.

```python
boro_centroid = (
    sales_pluto_combined
    .groupby("BOROUGH")[["latitude","longitude"]]
    .median()
)

mask = sales_pluto_combined["latitude"].isna()

sales_pluto_combined.loc[mask, "latitude"] = (
    sales_pluto_combined.loc[mask, "BOROUGH"]
    .map(boro_centroid["latitude"])
)

sales_pluto_combined.loc[mask, "longitude"] = (
    sales_pluto_combined.loc[mask, "BOROUGH"]
    .map(boro_centroid["longitude"])
)
```

As of now (January 22, 2026), the official CPI for January 2026 has not yet been published by the U.S. Bureau of Labor Statistics — it is scheduled for release on February 11, 2026.

Because the official January 2026 CPI number isn’t available yet, many analysts and modelers use an estimated or forecast value for that month when adjusting prices forward. For example, CPI forecast models (such as those from financial forecast centers) project the U.S. CPI-U to be around 325.17 (1982-84=100) for January 2026.

```python
sales_pluto_cpi_combined['ADJUSTED SALE PRICE'] = sales_pluto_cpi_combined['SALE PRICE'] * (325.17 / sales_pluto_cpi_combined['CPI'])
```
